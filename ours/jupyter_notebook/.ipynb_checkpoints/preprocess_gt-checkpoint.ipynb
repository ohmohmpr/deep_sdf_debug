```python
from bbox.metrics import iou_3d
from kiss_icp.tools.utils_class import BoundingBox3D
from typing import List
from scipy.spatial.transform import Rotation as R
import numpy as np
import pandas as pd
import copy
from pathlib import Path
```

```python
# different id because using filter
annotations_and_detections = np.load("../results/annotations_and_detections/002/002046.npy", allow_pickle='TRUE').item()
new_one = {}
save_new_annotations = Path("../../sandbox/Deep_SDF_Debug/data/P04/gt/002/002046.npy")
```

```python
cleaned_annotation = {}
for frame_id in annotations_and_detections:
    
    frame = annotations_and_detections[frame_id]
    
    new_one[frame_id] = {}
    print("frame_id", len(frame))
    print("frame_id", len(frame))
    for track_uuid, anno_and_det in frame.items():
        # print("track_uuid", track_uuid)
        # if track_uuid == '7c5f2512-f968-4dc3-867f-2d8fc851868b':
        #     print("frame_id", frame_id)
        #     print("track_uuid", track_uuid)
        new_one[frame_id][track_uuid] = {}
    
        annotation = anno_and_det[1]
        x = annotation.dst_SE3_object.translation[0]
        y = annotation.dst_SE3_object.translation[1]
        z = annotation.dst_SE3_object.translation[2]
        length =  annotation.length_m
        width =  annotation.width_m
        height =  annotation.height_m
        rot = annotation.dst_SE3_object.rotation

        new_one[frame_id][track_uuid]['x'] = x
        new_one[frame_id][track_uuid]['y'] = y
        new_one[frame_id][track_uuid]['z'] = z
        new_one[frame_id][track_uuid]['length'] = length
        new_one[frame_id][track_uuid]['width'] = width
        new_one[frame_id][track_uuid]['height'] = height
        new_one[frame_id][track_uuid]['rot'] = rot

np.save(save_new_annotations, new_one, allow_pickle=True)
# print(new_one)
```
